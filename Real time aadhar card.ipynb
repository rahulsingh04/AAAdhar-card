```python
import cv2
from PIL import Image

cam = cv2.VideoCapture(0)
detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
Id = input("enter the id number")
import pyttsx3
engine = pyttsx3.init()
rate = engine.getProperty('rate')
engine.setProperty('rate' ,150)

say =( '.......Camera is Opening  in    the   Background   ,  So   be   in   a   Right position')
print(say)
engine.say(say)

engine.runAndWait()
sampleNum = 0
while (True):
    ret, img2 = cam.read()
    gray = cv2.cvtColor(img2, cv2.COLOR_BGRA2GRAY)
    faces = detector.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(img2, (x, y), (x +  w, y + h), (255, 0, 0), 2)

        # incrementing sample number 
        sampleNum = sampleNum + 1
        # saving the captured face in the dataset folder
        s=cv2.imwrite(
            #"F:/image/user.jpg",
            'F:/ID/id_card.jpg',
            gray[y:y + h, x:x + w])

        cv2.imshow('frame', img2)
    # wait for 100 miliseconds
    if cv2.waitKey(100) & 0xFF == ord('q'):
        break
    # break if the sample number is morethan 20
    elif sampleNum > 1:
         break
cam.release()
cv2.destroyAllWindows()


import cv2
#font = cv2.FONT_HERSHEY_SIMPLEX
font =  cv2.FONT_HERSHEY_PLAIN
img=cv2.imread('id4.png')

gim=Image.open("F:/ID/id_card.jpg")

#img = Image.open('id2.png')
#gim=Image.open("F:/image/user.jpg")

n=input("Enter the name:-")

cv2.putText(img, n  , (220 , 138) , font , 1 ,
                       (0,0,0),1,cv2.LINE_AA)
i = input('date of birth:-')


cv2.putText(img, i  , (252 , 158) , font , 1,
                       (0,0,0),1 ,cv2.LINE_AA)
f = input("Enter the gender of  the person-")

cv2.putText(img, f  , (212 , 179) , font , 1 ,
                       (0 ,0,0),1 ,cv2.LINE_AA)

f = input("Date of expire:-")

cv2.putText(img, f  , (260 , 201) , font , 1,
                       (0 ,0,0),1 ,cv2.LINE_AA)


f = input("ENTER your height:-")

cv2.putText(img, f  , (220 , 222) , font , 1,
                       (0 ,0,0),1 ,cv2.LINE_AA)

f = input("ENTER your eye colour:-")

cv2.putText(img, f  , (252 , 242) , font , 1,
                       (0 ,0,0),1 ,cv2.LINE_AA)


f = input("ENTER your Distiguishmark:-")

cv2.putText(img, f  , (285, 265) , font , 1,
                       (0 ,0,0),1 ,cv2.LINE_AA)


cv2.imwrite('F:/ID/chacha.jpg',img)
arav=Image.open("F:/ID/chacha.jpg")

gim=Image.open("F:/ID/id_card.jpg")

arav.paste(gim,(360,50))


cv2.waitKey(0)

arav.save("F://ID/result.jpg")
arav



```
